```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
%config InLineBackend.figure_format = 'KKX'
%matplotlib inline
```

```python
data = pd.read_csv('RESULT-4.csv')
data
```

```python
msk = np.random.rand(len(data)) < 0.8
train = data[msk]
test = data[~msk]
train
```

```python
poly = PolynomialFeatures(degree=2)
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['year','gdp']])
train_y = np.asanyarray(train[['value']])
train_x
```

```python
x_poly = poly.fit_transform(train_x)
x_poly
```

```python
reg = LinearRegression()
reg.fit(x_poly,train_y)
```

```python
reg.predict(x_poly)
```

```python
plt.scatter(train.year, train.value, color='blue')
plt.plot(train_x,reg.predict(x_poly),color = 'red')
```

```python
from sklearn.metrics import r2_score
test_x = np.asanyarray(test[['year','gdp']])
test_y = np.asanyarray(test[['value']])
test_y_ = reg.predict(poly.fit_transform(test_x))
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```
