```python
cdf = df[['data','gdp','val']]
cdf.head(60)
```

```python
plt.scatter(cdf.gdp, cdf.data,  color='blue')
plt.xlabel('gdp')
plt.ylabel('data')
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]

```

```python
plt.scatter(train.data, train.val,  color='red')
X = np.arange(-5.0, 5.0, 0.1)
Y = 1-4/(1+np.power(3, X-2))
plt.plot(X,Y) 
plt.show()
plt.xlabel('data')
plt.ylabel("val")
plt.show()
```

```python
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['data']])
train_y = np.asanyarray(train[['val']])
regr.fit (train_x, train_y)

print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['data']])
test_y = np.asanyarray(test[['val']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```
